```python
from IPython.display import display, Markdown
import re
import json
import IPython

def get_previous_markdown():
    """Extracts content between and from notebook JSON"""
    try:
        shell = IPython.get_ipython()
        if shell is None:
            return "Error: Not in an interactive IPython environment."

        notebook_path = shell.config.get('IPKernelApp', {}).get('connection_file', '').replace('connection.json', '/home/aliko/APPs/AuditMaster/AuditMaster/Untitled-1.ipynb')
        if not notebook_path:
            return "Error: Could not find notebook path."

        with open(notebook_path, 'r', encoding='utf-8') as f:
            notebook_content = json.load(f)

        for cell in notebook_content['cells']:
            if cell['cell_type'] == 'markdown':
                content = ''.join(cell['source'])
                if '' in content:
                    pattern = r'(.*?)'
                    match = re.search(pattern, content, re.DOTALL)
                    if match:
                        return match.group(1).strip()
        return "Error: No Markdown input found with markers."

    except FileNotFoundError:
        return "Error: Notebook file not found."
    except KeyError as e:
        return f"Error: KeyError: {e}"
    except Exception as e:
        return f"Error: {str(e)}"

# Usage
markdown_input = get_previous_markdown()
display(Markdown(f"### Captured Markdown:\n{markdown_input}"))
```

```python
from IPython.display import display, Markdown
import json
import re

def get_last_markdown():
    """Get content from the last Markdown cell containing <!-- INPUT --> markers"""
    notebook_path = !jupyter notebook list | grep -oP '(?<=/tree/).*'
    if not notebook_path:
        return "Error: Could not find notebook path"
    
    try:
        with open(notebook_path[0].strip(), 'r', encoding='utf-8') as f:
            notebook = json.load(f)
        
        # Find all Markdown cells with markers
        markdown_cells = [
            ''.join(cell['source'])
            for cell in notebook['cells']
            if cell['cell_type'] == 'markdown' 
            and '<!-- INPUT -->' in ''.join(cell['source'])
        ]
        
        if markdown_cells:
            # Extract content between markers
            content = re.search(r'<!-- INPUT -->(.*?)<!-- INPUT -->', 
                              markdown_cells[-1], re.DOTALL)
            return content.group(1).strip() if content else "No marked content found"
        return "No Markdown cells with markers found"
    
    except Exception as e:
        return f"Error: {str(e)}"

# Usage
markdown_content = get_last_markdown()
display(Markdown(f"**Captured Markdown:**\n{markdown_content}"))
```
